```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("FuelConsumption.csv")
```

```python
df.head()
```

```python
df["FUELTYPE"].value_counts()
```

```python
df_plot = df[df["FUELTYPE"]==4]
plt.scatter(x=df_plot["FUELCONSUMPTION_COMB"], y=df_plot["CO2EMISSIONS"])
```

```python
df_fuels = [df[df["FUELTYPE"]==f] for f in range(1, 5)]
```

```python
clf_fuels = [LinearRegression().fit(dff[["ENGINESIZE", "CYLINDERS", "FUELTYPE", "FUELCONSUMPTION_COMB_MPG",]], 
                                    dff["CO2EMISSIONS"]) for dff  in df_fuels]
```

```python
df["FUELTYPE"] = df["FUELTYPE"].astype('category') \
                               .cat.rename_categories(list(range(1, 1 + len(df["FUELTYPE"].unique())))) \
                               .astype(int)
```

```python
df_lr = df[["ENGINESIZE", "CYLINDERS", "FUELTYPE", "FUELCONSUMPTION_COMB_MPG", "CO2EMISSIONS"]]
```

```python
df_lr.head()
```

```python
rand_idx = np.random.rand(len(df_lr)) < 0.9
training_set = df_lr[rand_idx]
test_set = df_lr[~rand_idx]
```

```python
clf = LinearRegression().fit(training_set[["ENGINESIZE", "CYLINDERS", "FUELTYPE", "FUELCONSUMPTION_COMB_MPG",]], 
                             training_set["CO2EMISSIONS"])
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
r2_score(test_set["CO2EMISSIONS"], 
               clf.predict(test_set[["ENGINESIZE", "CYLINDERS", "FUELTYPE", "FUELCONSUMPTION_COMB_MPG",]]), 
              )
```

```python
N = np.random.randint(0, len(df_lr))
fuel = df_lr.iloc[N, 2]
print(df_lr.iloc[N,:])
```

```python
clf_fuels[fuel-1].predict([list(df_lr.iloc[N,:-1])])
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([('poly', PolynomialFeatures(degree=2)), ('clf', LinearRegression())])
```

```python
clf = pipe.fit(training_set[["ENGINESIZE", "CYLINDERS", "FUELTYPE", "FUELCONSUMPTION_COMB_MPG",]], 
                             training_set["CO2EMISSIONS"])
```

```python
clf.score(test_set[["ENGINESIZE", "CYLINDERS", "FUELTYPE", "FUELCONSUMPTION_COMB_MPG",]], 
                             test_set["CO2EMISSIONS"])
```

```python
sigma_fn = lambda z, a, b: 1/(1+np.exp(-a*(z-b)))
```
