```python
make, model, year = "volkswagen", "passat", 2006
```

```python
df_car = df[(df['VehicleModel']==model.upper()) & \
            (df['VehicleMake']==make.upper()) & \
            (df['YearOfBirth']==year)]
df_idx = df_car.index[0]
df_car_truncated = df.iloc[df_idx,:].to_frame()
df_car_failure = df_car_truncated.transpose()[df_car.columns[df_car.columns.str.contains("%")]]
failure = float(df.iloc[df_idx, 7])


df_car_failure = df_car_failure.transpose().sort_values(by=df_idx, ascending=False)


failure_desc_list = df_car_failure.index[2:7].tolist()
failure_percentage = df_car_failure.iloc[2:7, 0]

failures_body = ""
for d, p in zip(failure_desc_list, failure_percentage):
    failures_body = failures_body + "{}: {}\n".format(d, p)

return_body = "<h1>{} {} {}: {}% fails<h1>".format(make, model, year, failure)
return_body += failures_body
```

```python
df_plot = df[["VehicleMake", "PASS %"]].groupby(["VehicleMake"]).agg(['count', 'mean', 'std'])
```

```python
plt.scatter(x=df_plot['count'], y=df_plot['mean'])
plt.xlabel("Number of Vehicles")
plt.ylabel("Pass Rate")
```

```python
sns.regplot(x="count", y="mean", data=df_plot)
```

```python
sns.distplot(df_plot["mean"])
```

```python
df_plot.columns = ["count", "mean", "std"]
```

```python
df_honest = df_plot[df_plot['count'] > 50]
df_honest['max_pass'] = df_honest['mean'] + df_honest['std']
df_honest.sort_values(by='max_pass', ascending=True, inplace=True)
```

```python
fig = plt.figure(figsize=(10, 25))
ax = df_honest['mean'].plot(kind='barh',xerr=df_honest['std'])
ax.xaxis.tick_top()
ax.xaxis.set_label_position('top') 
plt.title("NCT Pass Rate based on the Make of the Vehicle (2018 data)")
plt.tight_layout()
plt.savefig("/Users/aspiela/Desktop/nct_pass_rate.png")
```
