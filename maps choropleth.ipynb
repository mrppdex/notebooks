```python
df_incidents = pd.read_csv('https://ibm.box.com/shared/static/nmcltjmocdi8sd5tk93uembzdec8zyaq.csv')

```

```python
df_small = df_incidents.loc[0:1000,:]
df_small.shape
```

```python
sf_map = folium.Map(location=(37.773972, -122.431297), zoom_start=12.)
```

```python
from folium.plugins import MarkerCluster
```

```python
markers = MarkerCluster().add_to(sf_map)
```

```python
for lat, lon, lab in zip(df_small["X"], df_small["Y"], df_small["Category"]):
    folium.Marker(location=(lon, lat), popup=lab).add_to(markers)
```

```python
world_geo = r'world_countries.json'
```

```python
world_map = folium.Map(location=(0, 0), zoom_start=2, tiles='Mapbox Bright')
```

```python
threshold_scale = np.linspace(
                        1,
                        df_can.Total.max()+1,
                        6
                    )
```

```python
world_map.choropleth(
    geo_data = world_geo,
    data = df_can,
    columns = ["Country", "Total"],
    key_on = "feature.properties.name",
    threshold_scale = threshold_scale,
    fill_color='RdYlGn', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Immigration to Canada'
)
```
