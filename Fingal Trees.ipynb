```python
df = pd.read_csv("fingaltrees.csv")
```

```python
df.head()
```

```python
df_trees = df.dropna().reset_index(drop=True)
```

```python
df_trees.head()
```

```python
import folium
```

```python
!conda install -c conda-forge geopy --yes
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
```

```python
address = "Fingal, Dublin, Ireland"

geolocator = Nominatim(user_agent="my_app")
location = geolocator.geocode(address)
longitude = location.longitude
latitude = location.latitude
```

```python
fingal_map = folium.Map(location=(latitude, longitude), 
                        zoom_start=10,
                        tiles="Stamen Toner",
                       )
#fingal_map
```

```python
label = folium.Tooltip("Fingal")
folium.Marker((latitude, longitude), tooltip=label).add_to(fingal_map)
```

```python
fingal_map
```

```python
df_small = df_trees.head(1000)
```

```python
from folium.plugins import MarkerCluster

tree_cluster = MarkerCluster().add_to(fingal_map)
```

```python
def print_row(row):
    #print(f"{row['Common_Name']} at ({row['Lat']}, {row['Long']})")
    lat = row['Lat']
    lon = row['Long']
    tree_name = folium.Popup(row['Common_Name'], parse_html=True)
    folium.CircleMarker(
                (lat, lon), 
                radius=5,
                color='green',
                fill=True,
                popup=tree_name
    ).add_to(tree_cluster)
    
_ = df_small[['Lat', 'Long', 'Common_Name']].apply(print_row, axis=1)
```

```python
fingal_map
```

```python
filename = "fingal_trees.html"

with open(filename, "w") as f:
    content = f"<!doctype html><html><head><title>Trees in co. Fingal</title><body>{fingal_map._repr_html_()}</body></html>"
    f.write(content)

#fingal_map._repr_html_()
```
